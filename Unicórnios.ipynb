```python
#Importar as libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
#Ler os dados
base_dados = pd.read_csv('unicorns till sep 2022.csv')

#Verificar Dimensão
base_dados.shape
```

```python
#Primeiros registros
base_dados.head()
```

```python
#Colunas
base_dados.columns
```

```python
#Verificar Tipo do informação
base_dados.info()
```

```python
#Insere a coluna Id
base_dados.insert(0, 'Id', range(1, 1 + len(base_dados))) #Posição da coluna em que id será inserido, Nome da coluna e os dados que ele irá conter

#Definindo Id como índice
base_dados.set_index('Id', inplace=True)
```

```python
base_dados
```

```python
#Verificando campos Nulos e quantos campos nulos tem
base_dados.isnull().sum()
```

```python
#Visualização gráfica de números nulos
plt.figure(figsize=(10,6))
plt.title('Quantidade de campos nulos')
sns.heatmap(base_dados.isnull());
```

```python
#Análise de valores únicos
base_dados.nunique()
```

```python
#Rank de valores únicos
base_dados['Industry'].value_counts()
```

```python
#Gerando um gráfico de barras para visualização
plt.figure(figsize=(10,6))
plt.title('Quantidade de Empresas por Setor')
plt.xlabel('Setor')
plt.ylabel('Quantidade de Empresas')
plt.xticks(rotation=90)
sns.barplot(base_dados['Industry'].value_counts(), palette=['red', 'blue'])
```

```python
#Verificando o percentual de geração de unicornios de cada país
pct = base_dados['Country'].value_counts(normalize=True)[:10]
pct
```

```python
plt.figure(figsize=(15,10))
plt.title('Percentual de geração de unicornios de cada país')
plt.xticks(rotation=90)
plt.pie(pct, labels=pct.index, autopct='%1.1f%%', startangle=90);
```

```python
#Conversão de campo
base_dados['Date Joined'] = pd.to_datetime(base_dados['Date Joined'])

#Verificando a conversão
base_dados['Date Joined'].head()

#Extraindo ano e mês
base_dados['Year Joined'] = base_dados['Date Joined'].dt.year
base_dados['Month Joined'] = base_dados['Date Joined'].dt.month

#Verificando a nova coluna
base_dados.head()
```

```python
#Tabela analítica
analise_agr = base_dados.groupby(by=['Country', 'Year Joined', 'Month Joined']).count().reset_index()
analise_agr
```

```python
#Tabela analítica do Brasil
analise_agr.loc[analise_agr['Country'] == 'Brazil']
```

```python
#Retirando o sifrão dos valores de mercado
base_dados['Valuation ($B)'] = base_dados['Valuation ($B)'].str.replace('$', '')
```

```python
#Transformando os valores de valuation em números
base_dados['Valuation ($B)'] = base_dados['Valuation ($B)'].astype(float)
```

```python
#Tabela que soma o valor de cada setor em cada país
tbl_anl_cnt = base_dados.groupby(by=['Country', 'Industry'])['Valuation ($B)'].sum().reset_index()
tbl_anl_cnt
```

```python
#Tabela que soma o valor de todos os setor de cada país
tbl_anl_cnt = base_dados.groupby(by=['Country'])['Valuation ($B)'].sum().reset_index()
tbl_anl_cnt
```

```python
#Criando um gráfico para representar
plt.figure(figsize=(15,10))
plt.title('Valor de Unicornios por País')
plt.plot(tbl_anl_cnt['Country'], tbl_anl_cnt['Valuation ($B)'])
plt.xticks(rotation=90);
```
